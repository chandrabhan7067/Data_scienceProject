```python
import pandas as pd
```

```python
data = pd.read_csv("collegePlace.csv")
```

```python
data.head()
```

```python
# In our dataset there are two problem,that Gender and Stream is in the form of string so we need to replace it with integer(categorical) value
```

```python
data['Gender'].replace(['Male','Female'],[1,0],inplace=True)
```

```python
data.head()
```

```python
data['Stream'].unique()
```

```python
data['Stream'].replace({'Electronics And Communication':0,'Computer Science':1,'Information Technology':2,'Mechanical':3,'Electrical':4,'Civil':5},inplace = True)
data.head()
```

```python
#After this all let's see is there null value available or not
data.isnull().sum()
```

```python
#Now we will have to separate dependent and independent data 
y = data['PlacedOrNot']     #Dependent dataset
x = data.drop(['PlacedOrNot'],axis=1)    #Independent dataset
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.1)     #Here we are spliting dataset into train(90%) and test(10%) for finding the correct result
```

```python
X_train
```

```python
#we can do this with the help of logisticregression ðŸ˜€
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train,Y_train)
clf.score(X_test, Y_test)
```

```python
#We have one another option XGBClassifier which give us a higher accuracy ðŸ˜€
#use to solve the classification problem
from xgboost import XGBClassifier
clf = XGBClassifier()
clf.fit(X_train, Y_train)
clf.score(X_test, Y_test)
```
