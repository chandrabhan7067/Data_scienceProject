```python
# our label is present in the form of text and we need to convert it into numrical data because machine can easily understand the numerical data
from sklearn.preprocessing import LabelEncoder
Y = LabelEncoder().fit_transform(Y)
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
#Most of the times, your dataset will contain features highly varying in magnitudes, units and range. But since, most of the machine learning algorithms use Eucledian distance between two data points in their computations. We need to bring all features to the same level of magnitudes. This can be achieved by scaling. This means that youâ€™re transforming your data so that it fits within a specific scale, like 0â€“100 or 0â€“1.
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

#If we will use the fit method on our test data too, we will compute a new mean and variance that is a new scale for each feature and will let our model learn about our test data too. Thus, what we want to keep as a surprise is no longer unknown to our model and we will not get a good estimate of how our model is performing on the test (unseen) data which is the ultimate goal of building a model using machine learning algorithm.
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, Y_train)
classifier.score(X_test, Y_test)
```
